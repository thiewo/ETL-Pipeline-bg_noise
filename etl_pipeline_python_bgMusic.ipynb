```python
import pandas as pd

df = pd.read_csv('background_noise_focus_dataset.csv')

print(df.head())
print(df.shape)
print(df.columns.tolist())
```

```python
df = df.dropna()
df.columns = df.columns.str.lower().str.replace(' ', '_')

print("Coloane:", df.columns.tolist())
print("Rânduri rămase:", len(df))
```

```python
print(df.describe())
col_text=df.select_dtypes(include='object').columns[0]
print(col_text)
```

```python
import sqlite3

# Create a SQLite database connection
conn = sqlite3.connect('my_database.db')

# Save the DataFrame to a SQL table
df.to_sql('my_table', conn, if_exists='replace', index=False)

# Close the connection
conn.close()

print("DataFrame successfully saved to my_database.db as 'my_table'.")
```
